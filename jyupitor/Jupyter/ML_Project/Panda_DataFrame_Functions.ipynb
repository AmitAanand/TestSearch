# Reading CSV file in Dataframe

```python
train_data=pd.read_csv('C:\\Users\\kamit\\Desktop\\DataScience\\Panda\\train.csv')
df_train_data=pd.DataFrame(train_data)

df_train_data.head()
```

# Creating dummy variables 

```python
dummy_col_sex=pd.get_dummies(df_train_data['Sex'],prefix='sex')
dummy_col_sex.head()
```

**Joining the dummy Variables**

```python
col_name=df_train_data.columns.values.tolist()   # Taking column headings in the list
col_name.remove('Sex')                           # remove Sex column from the list
df_train_data[col_name].join(dummy_col_sex).head()      # Display top 5 records joining dunmmy variables 
```

```python
df_train_data.head()             # Original value remains same
```

# Grouping the data – aggregation, filtering, and transformation 

```python
group_col_data= df_train_data.groupby('Sex')
group_col_data.get_group('male').head()
```

```python
group_col_data= df_train_data.groupby(['Sex','Embarked'])
group_col_data.get_group(('female','Q')).head()
```

**Aggregate**

```python
import numpy as np

group_col_data_1=df_train_data.groupby(['Sex','Embarked'])

group_col_data_1.get_group(('male','Q')).head()
```

*Several Functions to all the Columns at the same time*

```python
group_col_data_1.aggregate({'Fare':np.sum,'Age':np.mean})
```

```python
group_col_data_1.aggregate([np.sum,np.mean,np.std])
```
