```python
BatchA_Score = pd.DataFrame({
    "Name": ['Alex','Mark','Zack'],
    "Score": [88,79,89],
    "Subject": ['Python','Java','C++']
})


BatchB_Score = pd.DataFrame({
    "Name": ['Martin','Rojer','Kelly'],
    "Score": [89,78,90],
    "Subject": ['Python','Java','C++'],
    "Total":[100,100,100]
})
```

```python
BatchB_Score
```

```python
concat1=pd.concat([BatchA_Score,BatchB_Score],sort=True,ignore_index=True)
concat1
```

```python
concat2=pd.concat([BatchA_Score,BatchB_Score],keys=["BatchA","BatchB"],sort=False)
concat2
```
